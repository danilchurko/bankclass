```python
import numpy as np
import pandas as pd
from pprint import pprint

import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
```

# Import data

```python
# Some missing values were represented as inf and -inf
missing_values = ['inf', '-inf']
```

```python
train = pd.read_csv('Train.csv', na_values=missing_values)
test = pd.read_csv('Test.csv', na_values=missing_values)
```

```python
print(f'Train: {train.shape}')
print(f'Test: {test.shape}')
```

```python
train.head()
```

```python
test.head()
```

# Split and drop for train

```python
X = train.drop(['id','column_1','column_2','column_3','column_4','column_5'],axis=1)
Y = train['column_5']
```

```python
X.head()
```

```python
Y.head(10)
```

# Handling missing values

```python
missing = X.isnull().sum()
missing.sum()
```

```python
X.fillna(X.mean(skipna=True), inplace=True)
```

```python
missing = X.isnull().sum()
missing.sum()
```

# Random Forest | sklearn

```python
n_estimators = [int(x) for x in np.linspace(start = 2, stop = 20, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(1, 20, num = 20)]
max_depth.append(None)
min_samples_split = [2, 5, 10, 20]
min_samples_leaf = [1, 2, 3, 5]
bootstrap = [True, False]
```

```python
# Creating the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Using the random grid to search for best hyperparameters
rforest = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation
rf_random = RandomizedSearchCV(estimator = rforest, param_distributions = random_grid, n_iter = 15,
                               cv = 5, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs = -1)
```

```python
rf_random.fit(X, Y.values)
params = rf_random.best_params_
pprint(params)
```

 Fitting Random Forest Classifier with the best parameters

```python
rforest = RandomForestClassifier(**params)
rforest.fit(X , Y)
```

# Measuring performance

```python
Y_train_predicted = rforest.predict(X)
```

```python
cmrforest = confusion_matrix(Y.values, Y_train_predicted)

plt.figure(figsize=(6,5))
sn.heatmap(cmrforest ,annot=True , fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
rforest.score(X,Y)
```

# Make predictions

```python
X_test = test.drop(['id'],axis=1)
X_test.fillna(X_test.mean(skipna=True), inplace=True)
```

```python
Y_predicted = rforest.predict(X_test)
```

```python
index = np.arange(0, Y_predicted.shape[0])
data = {'id':index, 'column_5':Y_predicted}
df = pd.DataFrame(data=data)
df.head()
```
